# CodeSearchNet

```python
import random
import jsonlines
import os
from langdetect import detect

def is_english(text):
    try:
        return detect(text) == 'en'
    except:
        return False

def generate_random_select_codesearchnet_data(
    train,
    test,
    number,
    insturction,
    input_insturction,
    output_insturction,
    output_path,
):
    print("Start generate random select codesearchnet data...")
    samples = []
    random.shuffle(train)
    random.shuffle(test)

    max_input_length = 1024

    for idx in range(10000):
        if len(samples) == 5000:
            break
        if len(train[idx]["code_tokens"]) > max_input_length or len(test[idx]["code_tokens"]) > max_input_length :
            continue
        train_code_string = " ".join(train[idx]["code_tokens"]).strip()
        test_code_string = " ".join(test[idx]["code_tokens"]).strip()
        train_doc_string = " ".join(train[idx]["docstring_tokens"]).strip()
        test_doc_string = " ".join(test[idx]["docstring_tokens"]).strip()
        if not is_english(train_doc_string) or not is_english(test_doc_string):
            continue
        train_prompt = ""
        train_prompt += insturction
        train_prompt += input_insturction
        train_prompt += "\n"
        train_prompt += train_code_string
        train_prompt += "\n"
        train_prompt += output_insturction
        train_prompt += "\n"

        test_prompt = ""
        test_prompt += insturction
        test_prompt += input_insturction
        test_prompt += "\n"
        test_prompt += test_code_string
        test_prompt += "\n"
        test_prompt += output_insturction
        test_prompt += "\n"
        
        samples.append(
            {"prompt": train_prompt, "label": train_doc_string}
        )
        samples.append(
            {"prompt": test_prompt, "label": test_doc_string}
        )

    # select 'number' random train samples
    samples = random.sample(samples, number)
    
    with jsonlines.open(
        os.path.join(output_path, "random_samples_" + str(number) + ".jsonl"), "w"
    ) as f:
        f.write_all(samples)

    print("Finish generate random select codesearchnet data...")
    
    return samples
```

```python
result_save_path = "../../../data/songda/code_search_net_java"
train = []
with jsonlines.open(
    f"{result_save_path}/train.jsonl"
) as f:
    for i in f:
        train.append(i)
test = []
with jsonlines.open(
    f"{result_save_path}/test.jsonl"
) as f:
    for i in f:
        test.append(i)

print("train size: ", len(train))
print("test size: ", len(test))

number = 5000
system_prompt = "You are a helpful, respectful and honest assistant with a deep knowledge of code and software design. Always answer as helpfully as possible. If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct. If you don't know the answer to a question, please don't share false information.\n"
insturction = (
    system_prompt + "Generate comment (summarization) for this code  \n"
)
input_insturction = " [input] "
output_insturction = " [output] "

```

```python
samples = generate_random_select_codesearchnet_data(
    train,
    test,
    number,
    insturction,
    input_insturction,
    output_insturction,
    result_save_path,
)

# print(samples[86])

print(len(samples))
```

# TL-CodeSum

```python
import random
import jsonlines
import os
from langdetect import detect

def is_english(text):
    try:
        return detect(text) == 'en'
    except:
        return False

def generate_random_select_tlcodesum_data(
    train,
    test,
    number,
    insturction,
    input_insturction,
    output_insturction,
    output_path,
):
    print("Start generate random select tlcodesum data...")
    samples = []
    random.shuffle(train)
    random.shuffle(test)
    max_input_length = 1024

    for idx in range(number):
        if len(samples) == number:
            break
        if len(train[idx]["code"]) > max_input_length or len(test[idx]["code"]) > max_input_length :
            continue
        train_code_string = train[idx]["code"].strip()
        train_doc_string = train[idx]["comment"].strip()
        test_code_string = test[idx]["code"].strip()
        test_doc_string = test[idx]["comment"].strip()
        if not is_english(train_doc_string) or not is_english(test_doc_string):
            continue
        train_prompt = ""
        train_prompt += insturction
        train_prompt += input_insturction
        train_prompt += "\n"
        train_prompt += train_code_string
        train_prompt += "\n"
        train_prompt += output_insturction
        train_prompt += "\n"

        test_prompt = ""
        test_prompt += insturction
        test_prompt += input_insturction
        test_prompt += "\n"
        test_prompt += test_code_string
        test_prompt += "\n"
        test_prompt += output_insturction
        test_prompt += "\n"

        
        samples.append(
            {"prompt": train_prompt, "label": train_doc_string}
        )
        samples.append(
            {"prompt": test_prompt, "label": test_doc_string}
        )
    samples = random.sample(samples, number)
    with jsonlines.open(
        os.path.join(output_path, "random_samples_" + str(number) + ".jsonl"), "w"
    ) as f:
        f.write_all(samples)

    print("Finish generate random select tlcodesum data...")
    
    return samples
```

```python
result_save_path = "../../../data/songda/tl_code_sum"
train = []
with jsonlines.open(
    f"{result_save_path}/train.json"
) as f:
    for i in f:
        train.append(i)

test = []
with jsonlines.open(
    f"{result_save_path}/test.json"
) as f:
    for i in f:
        test.append(i)
number = 5000
system_prompt = "You are a helpful, respectful and honest assistant with a deep knowledge of code and software design. Always answer as helpfully as possible. If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct. If you don't know the answer to a question, please don't share false information.\n"
insturction = (
    system_prompt + "Generate comment (summarization) for this code  \n"
)
input_insturction = " [input] "
output_insturction = " [output] "

samples = generate_random_select_tlcodesum_data(
    train,
    test,
    number,
    insturction,
    input_insturction,
    output_insturction,
    result_save_path,
)

print(len(samples))
for i in samples[10:]:
    print(i["prompt"])
    print(i["label"])
    print("=========================================")
```
